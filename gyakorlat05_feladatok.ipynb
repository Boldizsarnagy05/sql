1. Listázzuk az egyes csoportokat, és a csoportokba tartozó termékek darabszámát!
2. A lista jelenítse meg a végösszeget is!

````sql
SELECT 
CASE WHEN seged.kategoria IS NULL THEN 'Összesen' ELSE kategoria END as kategoria, seged.db
FROM(
SELECT
CASE WHEN keszlet < 3000 THEN 'olcsó'
WHEN keszlet > 5000 THEN 'drága'
ELSE 'közepes' END as kategoria, count(distinct TERMEKKOD) as db
FROM Termek
GROUP BY ROLLUP (CASE WHEN keszlet < 3000 THEN 'olcsó'
WHEN keszlet > 5000 THEN 'drága'
ELSE 'közepes' END)) as seged
````

````sql
SELECT SORSZAM, EGYSEGAR*MENNYISEG as ertek,
SUM(EGYSEGAR*MENNYISEG) OVER(PARTITION BY SORSZAM),
SUM(EGYSEGAR*MENNYISEG) OVER(ORDER BY SORSZAM, TERMEKKOD),
SUM(EGYSEGAR*MENNYISEG) OVER(ORDER BY SORSZAM ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW)
FROM Rendeles_tetel
ORDER BY SORSZAM
````
